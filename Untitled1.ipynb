```julia
using LinearAlgebra, Statistics, Plots
using Flux
```

```julia
const height = 25
const width = 25 # canvas dimensions
const box_height = 5
const box_width = 5

function draw_box(x, y)
    canvas = zeros(height, width)
    canvas[mod.(x:x+box_height, height).+1, mod.(y:y+box_width, width).+1] .= 5
#     canvas = reshape(canvas, (height, width, 1, 1))
    return Array{Float32}(canvas)
end
```

```julia
x,y = 7, 7
canvas = draw_box(x,y);
# heatmap(canvas)
```

```julia
W = 0.1*randn(height*width, 2);
V = 0.1*randn(2, height*width);
enc(x,W) = relu.(reshape(x, (1,height*width)) * W)
dec(x,V) = reshape(x * V, (height, width))
```

```julia
out = enc(canvas, W);
size(out)
```

```julia
pred_ = dec(randn(1,2), V)
heatmap(dec(enc((draw_box(10,12) - pred_), W), V))
```

```julia
heatmap(dec(enc(draw_box(10,10), W), V))
```
