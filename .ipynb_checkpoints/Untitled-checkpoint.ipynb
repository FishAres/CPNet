```julia
using LinearAlgebra, Statistics, Plots
using Flux
```

```julia
const height = 25
const width = 25 # canvas dimensions
const box_height = 5
const box_width = 5

canvas = zeros(height, width)
function draw_box(x, y)
    canvas = zeros(height, width)
    canvas[mod.(x:x+box_height, height).+1, mod.(y:y+box_width, width).+1] .= 5
    return Array{Float32,2}(canvas)
end
```

```julia
canvas = draw_box(10,5)
heatmap(canvas)
```

```julia
n_channels = 1
in_canvas = canvas;
using CuArrays
```

```julia
m = Chain(
    x -> reshape(x, (height*width)),
    Dense(height*width, 32, relu),
    Dense(32, 16, relu),
    LSTM(16, 32),
    x -> relu(x),
    Dense(32, height*width, relu),
    Dropout(0.2),
    x -> reshape(x, (height, width))
    ) 
```

```julia
loss(x,y) = Flux.mse(m(x), y)
opt = Descent(0.05)
ps = params(m);
```

```julia
x,y = 2,4
xs = [Tracker.data(draw_box(x+i, y+2i)) |> gpu for i in 1:500];
```

```julia
pred_ = m(randn(height, width))
losses = []
for i in 1:length(xs)
    data = xs[i]
    Flux.train!(loss, ps,[(data-pred_, data)], opt)
    Flux.truncate!(m)
    pred_ = m(data)
    append!(losses, loss(pred_, data).data)
end
```

```julia
pls = []
x,y = 10,1
push!(pls,heatmap(m(draw_box(x,y)).data))
push!(pls,heatmap(draw_box(x,y)))
plot(pls..., layout=2)
```

```julia
plot(losses)
```
