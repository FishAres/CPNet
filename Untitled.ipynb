```julia
using LinearAlgebra, Statistics, Plots
using Flux
```

```julia
const height = 25
const width = 25 # canvas dimensions
const box_height = 5
const box_width = 5

function draw_box(x, y)
    canvas = zeros(height, width)
    canvas[mod.(x:x+box_height, height).+1, mod.(y:y+box_width, width).+1] .= 5
    return Array{Float32,2}(canvas)
end
```

```julia
canvas = draw_box(10,5)
heatmap(canvas)
```

```julia
m = Chain(
    x -> reshape(x, (height*width)),
    Dense(height*width, 32, relu),
    Dense(32, 16, relu),
    LSTM(16, 32),
    x -> relu(x),
    Dense(32, height*width, relu),
    x -> reshape(x, (height, width))
    )
```

```julia
loss(x,y) = Flux.mse(m(x), y)
opt = Descent(0.06)
ps = params(m);
```

```julia
x,y = 2,4
xs = [Tracker.data(draw_box(x+i, y+2i)) for i in 1:5000];
```

```julia
pred_ = m(randn(height, width))
losses = []
for i in 1:length(xs)
    data = xs[i]
    Flux.train!(loss, ps,[(data-0.5*pred_, data)], opt)
    Flux.truncate!(m)
    pred_ = m(data)
    append!(losses, loss(pred_, data).data)
end
```

```julia
pls = []
x,y = 15,1
push!(pls,heatmap(m(draw_box(x,y)).data))
push!(pls,heatmap(draw_box(x,y)))
plot(pls..., layout=2)
```

```julia
plot(losses)
```

```julia
pred_ = m(randn(height, width))
pred_losses = []
for i in 1:100
    data = xs[i]

    pred_ = m(data-0.5*pred_)
    append!(pred_losses, loss(pred_, data).data)
end
```
